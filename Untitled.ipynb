```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

t_r = 1e-5
w_ee = 2
v_th = 20e-3
tau = 4e-3

f_in, tb_in = np.meshgrid(np.linspace(1., 1/t_r, 10), np.linspace(0, 30, 10))

def d_ts(f_in):
    return v_th / (w_ee * (1 - v_th / (tau * w_ee * f_in)) * np.square(f_in))

def ts(f_in):
    return -tau * np.log(1 - v_th / (w_ee * f_in * tau))

def d_f_out_d_f_in(f_in):
    return -1./np.square(ts(f_in) + t_r) * d_ts(f_in)

d_f_out = d_f_out_d_f_in(f_in)
d_tb_out = 0 #-tb_in / f_in * d_f_out + 1 / (f_in * (t_s + t_r))

plt.quiver(f_in, tb_in, np.log(-d_f_out), d_tb_out)
plt.show()
```
