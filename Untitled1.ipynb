```python
def f_out(p_in, f_in, w):
    denom = w_ee * f_in - alpha * np.power(f_in * p_in * tau_i + w, beta)
    x = v_th_e / (tau_e * denom)
    f = -1. / (tau_e * np.log(1 - x))
    f = np.where(denom <= 0, np.nan, f)
    f = np.where(x >= 1, np.nan, f)
    return f

def p_out(p_in, f_in, w):
    denom = w_ee * f_in
    x = (v_th_e + alpha * np.power(f_in * p_in * tau_i + w, beta) * tau_e) / (tau_e * denom)
    p =  -1. / (tau_e * np.log(1 - x))
    p = np.where(denom <= 0, np.nan, p)
    p = np.where(x >= 1, np.nan, p)
    return p
```

```python
tau_e = 4e-3
tau_i = 4e-3
v_th_e = 20e-3
w_ee = 40e-3
alpha = 1e-3
beta = 0.
```

```python
def p_diff(p_in, f_in, w=0):
    p_o = p_out(p_in, f_in, w)
    return p_o - p_in

def f_diff(p_in, f_in, w=0):
    f_o = f_out(p_in, f_in, w)
    return f_o - f_in
```

```python
def f_x_d_x(x):
    return -1 / (np.square(np.log(1 - x)) * (1 - x))

def f_i(f_e, p):
    return np.power(w_ei * f_e * tau_i * p, beta)

def f_i_d_f_e(f_e, p):
    return f_i(f_e, p) * beta / f_e

def f_i_d_p(f_e, p):
    return f_i(f_e, p) * beta / p

def f_e_d_f_e(f_e, p):
    f_inh = f_i(f_e, p)
    x = v_th_e / (tau_e * (w_ee * f_e + w_ie * f_inh))

    return 1 / tau_e * f_x_d_x(x) * np.square(x) * tau_e / v_th_e * (w_ee + w_ie * f_i_d_f_e(f_e, p))

def f_e_d_p(f_e, p):
    f_inh = f_i(f_e, p)
    x = v_th_e / (tau_e * (w_ee * f_e + w_ie * f_inh))
    
    return 1 / tau_e * f_x_d_x(x) * np.square(x) * tau_e / v_th_e * w_ie * f_i_d_p(f_e, p)

def p_d_f_e(f_e, p):
    f_inh = f_i(f_e, p)
    x = (v_th_e - w_ie * f_inh * tau_e) / (w_ee * f_e * tau_e)
    z = (w_ie * f_i_d_f_e(f_e, p)) / (w_ee * f_e) + (v_th_e - w_ie * f_inh * tau_e) / (w_ee * tau_e * np.square(f_e))
    return 1 / tau_e * f_x_d_x(x) * z
         
def p_d_p(f_e, p):
    f_inh = f_i(f_e, p)
    x = (v_th_e - w_ie * f_inh * tau_e) / (w_ee * f_e * tau_e)
    z = w_ie * tau_e * f_i_d_p(f_e, p)
    return 1 / tau_e * f_x_d_x(x) * z
         
def total_d_p(f_e, p):
    return p_d_p(f_e, p) + p_d_f_e(f_e, p) * f_e_d_p(f_e, p)

def total_d_f_e(f_e, p):
    return f_e_d_f_e(f_e, p) + f_e_d_p(f_e, p) * p_d_f_e(f_e, p)
```
